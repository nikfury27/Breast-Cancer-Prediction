#Splitting Dataset and Feature Scaling

```python
#splitting dataset into dependent and independent datasets
X = df.iloc[:,2].values.reshape(-1,1)
Y = df.iloc[:,1].values
```

```python
#split dataset into training (75%) and testing datasets(25%)
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25)
```

```python
#Scaling data(feature scaling)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
#print data
X_train
```

#Logistic Regression Model

```python
#build logistic regression classifier
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
Y_train = Y_train.astype(int)		#Y_train to numeric
Y_test = Y_test.astype(int)		        #Y_test to numeric
classifier.fit(X_train,Y_train)
classifier.get_params()
```

```python
#use of trained model to make predictions on data
predictions  = classifier.predict(X_test)
```

#Prediction Evaluation

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns
cm = confusion_matrix(Y_test,predictions)
print(cm)
sns.heatmap(cm,annot=True)
```

```python
#accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test,predictions))
```
